```python
import pandas as pd

ns_book7= pd.read_csv('ns_book7.csv',low_memory = False)

ns_book7.head()
```

```python

count_by_year = ns_book7['발행년도'].value_counts()
count_by_year
```

```python
count_by_year=count_by_year.sort_index()
count_by_year
```

```python
count_by_year = count_by_year[count_by_year.index <= 2030]
count_by_year
```

```python
import numpy as np
```

```python
def kdc_1st_char(no):
    if no is np.nan:
        return '-1'
    else:
        return no[0]

count_by_subject = ns_book7['주제분류번호'].apply(kdc_1st_char).value_counts()

count_by_subject
```

```python
#선그래프그리기

import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=100
```

```python
plt.plot(count_by_year.index, count_by_year.values)
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
#선모양 색상 바꾸기

plt.plot(count_by_year,marker='.',linestyle=":",color='r')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year,'*-g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
plt.plot(count_by_year,'*-g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947,2030,10))
for idx, val in count_by_year[::5].items():
    plt.annotate(val,(idx,val))
plt.show()
```

```python
plt.plot(count_by_year,'*-g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947,2030,10))
for idx, val in count_by_year[::5].items():
    plt.annotate(val,(idx,val),xytext=(0,), textcoords='offset points')
plt.show()
```

```python
#막대그래프 그리기
plt.bar(count_by_subject.index,count_by_subject.values)
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items():
    plt.annotate(val,(idx,val),xytext=(0,2), textcoords='offset points')
```

```python
plt.bar(count_by_subject.index,count_by_subject.values, width=0.7,color='b')
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items():
    plt.annotate(val,(idx,val),xytext=(0,2), textcoords='offset points',
    fontsize=8, ha='center',color='g')
```

```python
plt.barh(count_by_subject.index,count_by_subject.values, height=0.7,color='b')
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items():
    plt.annotate(val,(val,idx),xytext=(2,0), textcoords='offset points',
    fontsize=8, va='center',color='g')
```
